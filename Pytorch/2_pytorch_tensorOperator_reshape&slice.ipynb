```python
import torch
```

Reshape = ubah “bentuk wadah” <br>
view() → mirip reshape (tapi butuh memory contiguous) <br>
.flatten() → langsung jadi 1D <br>
Slice = ambil “isi sebagian”

```python
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape & view
my_torch = my_torch.reshape(2,5)
my_torch
```

```python
my_torch2 = torch.arange(15)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(3,-1)
my_torch2
```

```python
my_torch3 = torch.arange(10)
my_torch3
```

```python
my_torch3 = my_torch3.view(2,5)
my_torch3
```

```python
my_torch3[1] = 4321
my_torch3
```

```python
my_torch4 = torch.arange(10)
my_torch4
```

```python
my_torch4[1] = 1234
my_torch4
```

```python
#Slice
my_torch5 = torch.arange(10)
my_torch5 = my_torch5.reshape(5,2)
print(my_torch5)
print(my_torch5[:,0])
print(my_torch5[:,1])
print(my_torch5[:,1:])
```
