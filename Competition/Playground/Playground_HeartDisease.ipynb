```python
# CatBoost
cat = CatBoostRegressor(
    iterations=3000,
    learning_rate=0.02,
    depth=6,
    l2_leaf_reg=5,
    random_state=42,
    verbose=0
)

cat.fit(
    X_train, y_train,
    eval_set=(X_valid, y_valid),
    early_stopping_rounds=200
)



# LightGBM
lgbm = LGBMRegressor(
    n_estimators=3000,
    learning_rate=0.02,
    num_leaves=31,
    random_state=42
)

lgbm.fit(
    X_train, y_train,
    eval_set=[(X_valid, y_valid)],
    callbacks=[]
)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(
    n_estimators=3000,
    learning_rate=0.02,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    early_stopping_rounds=200   # ⬅️ pindah ke sini
)

xgb.fit(
    X_train, y_train,
    eval_set=[(X_valid, y_valid)],
    verbose=False
)
```
