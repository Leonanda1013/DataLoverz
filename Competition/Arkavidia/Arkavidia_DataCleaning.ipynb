```python
# kategorik
kolom_kategori = ['stasiun', 'critical', 'kategori']
```

```python
def clean_kolom_kategori(df, kolom_kategori):
    kolom = [c for c in kolom_kategori if c in df.columns]

    for col in kolom:
        df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .str.upper()
        )

    df[kolom] = df[kolom].astype('category')
    return df
```

```python

# 1. Pastikan kolom max bertipe numerik
ISPU_tanpa25['max'] = pd.to_numeric(ISPU_tanpa25['max'], errors='coerce')

# 2. Pastikan kategori bersih (hindari string 'NAN')
ISPU_tanpa25['kategori'] = (
    ISPU_tanpa25['kategori']
    .replace(['NAN', 'nan', 'None', 'NONE', ''], np.nan)
)

# 3. Fungsi penentuan kategori ISPU
def kategori_ispu(x):
    if pd.isna(x):
        return np.nan
    elif x <= 50:
        return 'BAIK'
    elif x <= 100:
        return 'SEDANG'
    elif x <= 200:
        return 'TIDAK SEHAT'
    elif x <= 300:
        return 'SANGAT TIDAK SEHAT'
    else:
        return 'BERBAHAYA'

# 4. Mask: hanya kategori NULL & max ADA
mask = ISPU_tanpa25['kategori'].isna() & ISPU_tanpa25['max'].notna()

# 5. Isi kategori NULL saja
ISPU_tanpa25.loc[mask, 'kategori'] = (
    ISPU_tanpa25.loc[mask, 'max']
    .apply(kategori_ispu)
)

# 6. (Opsional tapi disarankan) Jadikan category dtype
kategori_order = [
    'BAIK',
    'SEDANG',
    'TIDAK SEHAT',
    'SANGAT TIDAK SEHAT',
    'BERBAHAYA'
]

ISPU_tanpa25['kategori'] = pd.Categorical(
    ISPU_tanpa25['kategori'],
    categories=kategori_order,
    ordered=True
)

# 7. Cek hasil akhir
print("Jumlah kategori null:", ISPU_tanpa25['kategori'].isna().sum())
print(ISPU_tanpa25['kategori'].value_counts(dropna=False))
```
