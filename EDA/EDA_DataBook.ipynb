Diketahui bahwa pada nilai minimum Publishing Year terdapat minus, ini tidak mungkin terjadi , maka kita harus membetulkannya

```python
df['Publishing Year'].hist()
```

```python
df = df[df['Publishing Year'] > 1900]
```

```python
df.describe()
```

```python
df.hist(['Publishing Year'])
```

Cek apakah ada nilai null

```python
df.isna().sum()
```

Lakukan Hapus nilai null pada bookname karena kolom tersebut penting dan harusnya tidak boleh null

```python
df.dropna(subset = "Book Name", inplace = True)
```

```python
df.isna().sum()
```

Cek Duplikat

```python
df.duplicated().sum()
```

Cek nilai unik pada setiap kolom

```python
df.nunique()
```

```python
plt.hist(df['Publishing Year'])
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title('Distribution of Publishing Year')
plt.show()
```

```python
df['genre'].value_counts().plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Number of book')
plt.title('Distribution of Genre')
plt.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x="genre", y="Book_ratings_count", data = df)
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Box Plot Book Rating of Each Genre')
plt.show()
```

```python
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter Plot of Sale Price vs Units Sold")
plt.show()
```

```python
language_counts = df["language_code"].value_counts()
```

```python
plt.pie(language_counts, labels=language_counts.index, startangle=90, autopct="%1.1f%%")
plt.title("Distribution of Language Codes")
plt.show()
```
