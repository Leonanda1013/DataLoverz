
<h2> Deleting Redudant Columns






```python
data_raw.columns
```

```python
columns_to_keep = ['NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']
columns_to_drop = ['reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license','id']
```

1. Filtering Data

```python
df = data_raw[columns_to_keep]
```

```python
df
```

2. Dropping Columns

```python
data = data_raw.drop(columns=columns_to_drop)
```

```python
data
```

<h2> Rename Columns

1. Upper case & Lower case

```python
data.rename(columns ={'NAME' : 'Name'}, inplace = True)
```

```python
data.columns
```

```python
for i in data.columns :
  print(i.upper())
```

```python
new_col_names = []
for i in data.columns :
  new_col_names.append(i.lower())
  # append berfungsi untuk menambahkan data ke list
new_col_names
```

```python
new_dataframe_modifyd = data.copy()
new_dataframe_modifyd.columns = new_col_names
new_dataframe_modifyd
```
