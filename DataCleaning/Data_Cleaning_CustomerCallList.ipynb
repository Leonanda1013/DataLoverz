<h2> Drop Duplicates

```python
df.info()
```

```python
df.duplicated().value_counts()
```

```python
df = df.drop_duplicates()
```

<h2> Delete Coulumn

```python
df = df.drop('Not_Useful_Column',axis = 1)
```

```python
df
```

<h2> Strip

```python
df['Last_Name'] = df['Last_Name'].str.strip('12345./_')
df
```

<h2> Cleaning Phone Number

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
# dibaca kecuali a-zA-Z0-9
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
```

```python
df
```

<h2> Split Column

```python
df[['Street_Address', 'State', 'Zip_Code']]=df['Address'].str.split(',',expand=True)
```

```python
df
```

<h2> Standaring Column Value

```python
df['Paying Customer'].value_counts()
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('Y','Yes')
df['Paying Customer'] = df['Paying Customer'].replace('N','No')
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('N/a','')
```

```python
df = df.fillna('')
```

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Y','Yes')
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('N','No')
```

<h2> Filtering Column

```python
for x in df.index:
  if df.loc[x,'Do_Not_Contact'] == 'Yes':
    df.drop(x,inplace=True)
```

```python
df
```

```python
for x in df.index:
  if df.loc[x,'Phone_Number'] == '':
    df.drop(x,inplace=True)
```
