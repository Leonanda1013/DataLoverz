```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Membuat data
def sinusGenerator(amplitudo, frekuensi, tAkhir, theta):
  t = np.arange(0,tAkhir,0.1)
  y = amplitudo*np.sin(2*np.pi*frekuensi*t+theta)
  return t,y
t1,y1 = sinusGenerator(1,5,10,0)
t2,y2 = sinusGenerator(0.5,2,10,np.pi/2)
```

```python
# 2. Membuat Plot
plt.plot(t1,y1,'r', label='sinus')
plt.plot(t2,y2, 'b--', label='sinus 2')
plt.legend(loc='upper right')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')
plt.title('Grafik Sinusoidal')
plt.axis(False)
plt.grid(True)
# 3. menampilkan plot
plt.show()
```
